```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a sample dataset
# This could be your own data loaded from a file
data = {
    'Month': [1, 2, 3, 4, 5, 6, 7, 8],
    'Temperature': [22, 24, 23, 27, 28, 30, 29, 31]
}
df = pd.DataFrame(data)

# 2. Create the plot using seaborn.lineplot()
# Specify the columns for the x and y axes.
# Seaborn automatically connects the points with a line.
sns.lineplot(data=df, x='Month', y='Temperature', marker='o')

# 'marker="o"' is optional. It adds circles to clearly show where the data points are.

# 3. Add title and show the plot
plt.title('Monthly Temperature Trend')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.grid(True) # Optional: adds a grid for better readability
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a "long-format" DataFrame
# Notice the 'City' column, which will be used to separate the lines.
data = {
    'Month': [1, 2, 3, 4, 1, 2, 3, 4],
    'Temperature': [22, 24, 23, 27, 15, 14, 18, 17],
    'City': ['Butuan', 'Butuan', 'Butuan', 'Butuan', 'Surigao', 'Surigao', 'Surigao', 'Surigao']
}
df = pd.DataFrame(data)

# 2. Plot using the 'hue' parameter
# Seaborn creates a separate line for each unique value in the 'City' column.
sns.lineplot(data=df, x='Month', y='Temperature', hue='City', marker='o')

# 3. Add title and show the plot
plt.title('Monthly Temperature Comparison')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a "wide-format" DataFrame
data_wide = {
    'Month': [1, 2, 3, 4, 5, 6],
    'Butuan_Temp': [22, 24, 23, 27, 28, 30],
    'Surigao_Temp': [15, 14, 18, 17, 20, 22]
}
df_wide = pd.DataFrame(data_wide)

# 2. Call lineplot twice on the same axes
sns.lineplot(data=df_wide, x='Month', y='Butuan_Temp', label='Butuan', marker='o')
sns.lineplot(data=df_wide, x='Month', y='Surigao_Temp', label='Surigao', marker='s')

# 3. Add title, legend, and show the plot
plt.title('Monthly Temperature Comparison')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.legend() # This is needed to show the labels
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Create a sample dataset using a pandas DataFrame
# This "wide" format is very convenient for Matplotlib.
data = {
    'Month': [1, 2, 3, 4, 5, 6],
    'Butuan_Temp': [22, 24, 23, 27, 28, 30],
    'Surigao_Temp': [15, 14, 18, 17, 20, 22]
}
df = pd.DataFrame(data)

# 2. Plot the first line
# We explicitly select the columns from the DataFrame.
# The 'label' will be used for the legend.
plt.plot(df['Month'], df['Butuan_Temp'], label='Butuan', marker='o')

# 3. Plot the second line on the same canvas
# Matplotlib automatically adds it to the existing plot.
plt.plot(df['Month'], df['Surigao_Temp'], label='Surigao', marker='s')

# 4. Add details to make the plot informative
plt.title('Monthly Temperature Comparison (Matplotlib)')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.grid(True) # Adds a grid for better readability

# 5. Display the legend
# This function uses the 'label' arguments from the plot() calls.
plt.legend()

# 6. Show the final plot
plt.show()
```
